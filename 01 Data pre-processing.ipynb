```python
import os
import shutil
import numpy as np
import pandas as pd
```

```python
## The dataset can be downloaded from https://zenodo.org/records/4066594
csvdata = pd.read_csv('./reference_data.csv')
basedir = './Data_raw/'
disdir = './otolith_dataset/'

species = ['Alosa fallax', 'Ammodytes marinus', 'Ammodytes tobianus', 'Clupea harengus', 'Hyperoplus lanceolatus', 'Sprattus sprattus']
locations = ['Scheldt', 'BCS', 'DCS']

for spe in species:
   for loc in locations:
         os.mkdir(disdir + spe + '_' + loc)

for spefile in os.listdir(basedir):
    for speim in os.listdir(os.path.join(basedir + spefile)):
        imname = speim.split('.')[0]
        csvrow_matched = csvdata[csvdata['fish_id'] == imname[:17]]
        if len(csvrow_matched)>0:
            #print(csvrow_matched)
            # print(csvrow_matched['scientific_name'].iloc[0], csvrow_matched['sampling_location'].iloc[0])
            impath = os.path.join(basedir, spefile, speim)
            if os.path.isfile(impath):
                if imname[:17]==csvrow_matched['fish_id'].iloc[0] and spefile==csvrow_matched['scientific_name'].iloc[0]:
                    if impath[-6:] in ['L1.jpg', 'R1.jpg']:
                        print(impath)
                        print(csvrow_matched['scientific_name'].iloc[0] + '_' + csvrow_matched['sampling_location'].iloc[0])
                        shutil.copy(impath, os.path.join(disdir, csvrow_matched['scientific_name'].iloc[0] + '_' + csvrow_matched['sampling_location'].iloc[0]))
```

```python
import os

a = []  # list to store all file names

# Walk through the base directory (bottom-up to safely remove empty subfolders first)
for root, dirs, files in os.walk(disdir, topdown=False):
    # Collect file names
    for f in files:
        a.append(f)

    # Check subdirectories
    for d in dirs:
        dirpath = os.path.join(root, d)
        # If directory is empty, remove it
        if not os.listdir(dirpath):
            os.rmdir(dirpath)
            print(f"Removed empty folder: {dirpath}")

# Convert list to a set to keep unique file names
b = set(a)

print(f"Total files collected: {len(a)}")
print(f"Unique files: {len(b)}")
```
